```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
import datetime as dt
```

```python
file = 'C:/Users/mcarp/OneDrive/Documents/pm25compare.csv'
compare = pd.read_csv(file)
df = pd.DataFrame(compare)
```

```python
df['Safe Level (24 hrs)'] = 1464*[35]
df['Safe Level (1 year)'] = 1464*[12]
df['Good AQI'] = 1464*[50]
```

```python
df.head()
```

```python
dfHI=df[df['Site Name']=='Honolulu']
dfLA=df[df['Site Name']=='New Orleans / City Park']
dfFL=df[df['Site Name']=='Munro Street (Near-Road)']
dfOH=df[df['Site Name']=='GT Craig NCore']
print(dfHI.shape,dfLA.shape,dfFL.shape,dfOH.shape)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
plt.plot(range(1,367), 'Daily Mean PM2.5 Concentration', data=dfHI, marker='', color='blue', linewidth=2, label="Honolulu")
plt.plot(range(1,367), 'Daily Mean PM2.5 Concentration', data=dfLA, marker='', color='red', linewidth=2, label="New Orleans",alpha=0.8)
plt.plot(range(1,367), 'Daily Mean PM2.5 Concentration', data=dfFL, marker='', color='gray', linewidth=2, label="Tampa",alpha=0.4)
plt.plot(range(1,367), 'Daily Mean PM2.5 Concentration', data=dfOH, marker='', color='gray', linewidth=2, label="Cleveland", alpha=0.4)
plt.plot(range(1,367), 'Safe Level (24 hrs)', data=dfLA, marker='', color='olive', linewidth=2, linestyle='dashed', label="24 hr safe lvl")
plt.plot(range(1,367), 'Safe Level (1 year)', data=dfLA, marker='', color='orange', linewidth=2, linestyle='dashed', label="1 yr safe lvl")
plt.title('2020 Daily Mean PM 2.5 Concentration',size=18)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
plt.plot(range(1,367), 'DAILY_AQI_VALUE', data=dfHI, marker='', color='blue', linewidth=2, label="Honolulu")
plt.plot(range(1,367), 'DAILY_AQI_VALUE', data=dfLA, marker='', color='red', linewidth=2, label="New Orleans", alpha=0.8)
plt.plot(range(1,367), 'DAILY_AQI_VALUE', data=dfFL, marker='', color='gray', linewidth=2, label="Tampa",alpha=0.4)
plt.plot(range(1,367), 'DAILY_AQI_VALUE', data=dfOH, marker='', color='gray', linewidth=2, label="Cleveland", alpha=0.4)
plt.plot(range(1,367), 'Good AQI', data=dfLA, marker='', color='orange', linewidth=2, linestyle='dashed', label="Good vs. Moderate AQI")
plt.title('2020 Daily AQI Value')
plt.legend()
plt.show()
```
